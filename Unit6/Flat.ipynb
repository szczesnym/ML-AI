```python
import numpy as np
flat_price = 120000
flat_price_increase = 0.05
flat_years = 5
flat_months = flat_years * 12
```

```python
flat_price_in_years = flat_price * (1 + flat_price_increase)**np.arange(flat_years)
```

```python
flat_price_in_years
```

## Flat value after 5 years(yearly capitalization) 145860 

```python
flat_value_months = flat_price * (1 + flat_price_increase / 12)**np.arange(flat_months)
```

```python
flat_value_months
```

## Flat value after 5 years(monthly capitalization) 153364

```python
deposit_interest_year = 0.12
deposit_interest_month = deposit_interest_year /12
```

```python
deposit_interest_month
```

```python
payment = 1000
deposit_value = payment *np.arange(60) + payment * (1 + deposit_interest_month) ** np.arange(60)
```

```python
deposit_value
```

## Calculation when value of deposit will exceed value of flat, in 5 years, changing remmintance by 1 and calculating value of deposit

```python
payment = 1
deposit_payment_value = {}
while True:
    deposit_value = payment * np.arange(60) + payment * (1 + deposit_interest_month) ** np.arange(60)
    #print(f'Payment:{payment}, Deposit Value:{deposit_value[(59)]}')
    deposit_payment_value[payment] = (deposit_value[-1])
    if deposit_value[-1] > flat_value_months[-1]:
    #flat_price_in_years[(4)]:
        break
    payment = payment + 1
print(f'Final payment:{payment}, Final Deposit Value:%.2f'%deposit_value[-1])
```

# Minimal monthly payment to exceed flat value 2523

```python
import matplotlib.pyplot as plt
```

## Value of deposit in connection with value of monthly remittance

```python
plt.plot(list(deposit_payment_value.values()), label='Value of deposit in 60m')
plt.grid(axis='both')
plt.legend()
plt.ylabel('Deposit value')
plt.xlabel('Remittance')
```

```python
plt.plot(flat_value_months, label='Flat value')
plt.plot(deposit_value, label='Deposit Value')
plt.legend()
plt.title(f'Flat value %.f deposit value %.f'%(flat_value_months[-1], deposit_value[-1]))
plt.xlabel('Months')
plt.ylabel('Flat value')
```
