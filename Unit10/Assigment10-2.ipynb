```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
from scipy import stats
```

```python
df = pd.read_csv('titanic.csv')
```

```python
df.head()
```

```python
df.tail()
```

```python
df[df['Fare'].isnull()]
```

```python
mean = df['Fare'].mean()
mode = stats.mode(df['Fare'])
median = np.median(df['Fare'])
print(f'Mode: {mode}')
```

```python
q0 = np.quantile(df['Fare'], 0.00)
q1 = np.quantile(df['Fare'], 0.25)
q3 = np.quantile(df['Fare'], 0.75)
q4 = np.quantile(df['Fare'], 1.00)
print(f'Q0: {q0}')
print(f'Q1: {q1}')
print(f'Q2-mean: {mean}')
print(f'Q3: {q3}')
print(f'Q4: {q4}')
```

```python
variance = np.var(df['Fare'], ddof=1)
print(f'Variance: %.2f'%variance)
```

```python
deviance = np.std(df['Fare'], ddof=1)
print(f'Deviance: %.2f'%deviance)
```

```python
sns.boxenplot(data=df, y='Fare')
```

```python
plt.boxplot(df['Fare'])
plt.show()
```

```python
plt.figure(figsize=(10, 8))
sns.histplot(df['Fare'], bins=10)
plt.axvline(np.round(mean), color='r', label='Mean')
plt.axvline(np.round(median), color='b', label='Median')
plt.axvline(np.round(mode[0]), color='y', label='Mode')
plt.axvline(x = q0, color='g', label = 'Min')
plt.axvline(x = q1, color='w', label = 'Q1')
plt.axvline(x = q3, color='b', label = 'Q3')
plt.axvline(x = q4, color='y', label = 'Max')
plt.legend(loc='upper right')
plt.show()
```

```python
plt.figure(figsize=(15, 10))
sns.kdeplot(df['Fare'], fill=True)
plt.axvline(np.round(mean), color='r', label='Mean')
plt.axvline(np.round(median), color='b', label='Median')
plt.axvline(np.round(mode[0]), color='y', label='Mode')
plt.axvline(x = q0, color='g', label = 'Min')
plt.axvline(x = q1, color='w', label = 'Q1')
plt.axvline(x = q3, color='b', label = 'Q3')
plt.axvline(x = q4, color='y', label = 'Max')
plt.legend(loc='upper right')
plt.show()
```
