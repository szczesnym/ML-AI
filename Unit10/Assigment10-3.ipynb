```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
from pandas.core.interchange.dataframe_protocol import DataFrame
from scipy import stats
```

```python
x100 = np.linspace(0, 10, 100)
x1000 = np.linspace(0, 10, 1000)
x10000 = np.linspace(0, 10, 10000)
```

```python
x100
```

```python
distr_100 = stats.norm.pdf(x100, 3, 1) #np.random.normal(loc=3, scale=1, size=100)#
distr_1000 = stats.norm.pdf(x1000, 3, 1)
distr_10000 = stats.norm.pdf(x10000, 3, 1)#np.random.normal(loc=3, scale=1, size=10000)#
```

```python
distr_100
```

```python
plt.figure(figsize=(15, 10))
plt.plot(x100, distr_100, 'o')
plt.show()
plt.plot(x10000, distr_10000, '-')
plt.show()

sns.histplot(distr_100, bins=100, kde=True, color='orange')
plt.show()
sns.histplot(distr_10000, bins=100, kde=True, color='blue')
plt.show()
```

```python
mean_d_100 = distr_100.mean()
q0_d_100 = np.quantile(distr_100, 0.00)
q1_d_100 = np.quantile(distr_100, 0.25)
q3_d_100 = np.quantile(distr_100, 0.75)
q4_d_100 = np.quantile(distr_100, 1.00)
mode_d_100 = stats.mode(distr_100)
iqr_d_100 = stats.iqr(distr_100)
print(f'Q0_d_100: {q0_d_100}')
print(f'Q1d_100: {q1_d_100}')
print(f'Q2_d_100-mean: {mean_d_100}')
print(f'Q3d_100: {q3_d_100}')
print(f'Q4d_100: {q4_d_100}')
print(f'mode_d_100: {mode_d_100[0]}')
print(f'iqr_d_100: {iqr_d_100}')
```

```python
mean_d_10000 = distr_10000.mean()
q0_d_10000 = np.quantile(distr_10000, 0.00)
q1_d_10000 = np.quantile(distr_10000, 0.25)
q3_d_10000 = np.quantile(distr_10000, 0.75)
q4_d_10000 = np.quantile(distr_10000, 1.00)
mode_d_10000 = stats.mode(distr_10000)
iqr_d_10000 = stats.iqr(distr_10000)
print(f'Q0_d_10000: {q0_d_10000}')
print(f'Q1d_10000: {q1_d_10000}')
print(f'Q2_d_10000-mean: {mean_d_10000}')
print(f'Q3d_10000: {q3_d_10000}')
print(f'Q4d_10000: {q4_d_10000}')
print(f'mode_d_10000: {mode_d_10000[0]}')
print(f'iqr_d_10000: {iqr_d_10000}')
```

```python
labels=['Q0', 'Q1', 'Q3', 'Q4', 'Mean', 'Mode', 'IQR']
d100 = [q0_d_100, q1_d_100, q3_d_100, q4_d_100, mean_d_100, mode_d_100[0], iqr_d_100]
d10000 = [q0_d_10000, q1_d_10000, q3_d_10000, q4_d_10000, mean_d_10000, mode_d_10000[0], iqr_d_10000]
df=pd.DataFrame({'d100' : d100, 'd10000': d10000}, index=labels)
df['diff'] = df['d10000'] - df['d100']
```

```python
df
```

### Rznica w statystykach opisowych pomidzy prb 100 a 10000

```python
sns.barplot(data=df , y='diff',  x=df.index)
```

```python
print(f'd100 MIN:{distr_100.min()} MAX:{distr_100.max()}')
print(f'd10000 MIN:{distr_10000.min()} MAX:{distr_10000.max()}')
```

# Ale jak wemie si dane losowe, ktre maj rozkad normanly :) np.random.normal(loc=3, scale=1, size=10000)


```python
distr_100 = np.random.normal(loc=3, scale=1, size=100)
distr_10000 = np.random.normal(loc=3, scale=1, size=10000)
```

```python
plt.figure(figsize=(15, 10))
plt.plot(x100, distr_100, 'o', label='Prbka 100 dane')
plt.legend(loc='upper left')
plt.show()
plt.plot(x10000, distr_10000, '-', label='Prbka 10000 dane')
plt.legend(loc='upper left')
plt.show()

sns.histplot(distr_100, bins=100, kde=True, color='orange', label='Prbka 100 rozkad')
plt.legend(loc='upper left')
plt.show()
sns.histplot(distr_10000, bins=100, kde=True, color='blue', label='Prbka 10000 rozkad')
plt.legend(loc='upper left')
plt.show()
```

```python
print(f'd100 MIN:{distr_100.min()} MAX:{distr_100.max()}')
print(f'd10000 MIN:{distr_10000.min()} MAX:{distr_10000.max()}')
```
