```python
df_movies.groupby(pd.Grouper(key='release_date', freq='y')).agg({'revenue':'sum'})
```

```python
df_movies['release_year'] = df_movies['release_date'].dt.year.astype(int)
```

```python
df_graph_base = df_movies.pivot_table(values=['revenue', 'budget'], index='release_year', aggfunc='sum' ).reset_index()
df_graph_base
```

```python
fdf_graph_base = df_graph_base[(df_graph_base['release_year'] > 2009) & (df_graph_base['release_year'] < 2017) ]
```

```python
df_graph_base.groupby('revenue').sum()
```

```python
def million(x, pos):
        return '{:2.1f}B'.format(x*1e-9)
```

#### Zadanie punkt 2

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
axes = fig.add_axes([0,0,1,1])
#x = len(fdf_graph_base['release_year'])
#y = fdf_graph_base[ fdf_graph_base['revenue'].max() - fdf_graph_base['revenue'].min()]
axes.set_title('Budget and revenue in Bilions')
formatter = plt.FuncFormatter(million)
axes.yaxis.set_major_formatter(formatter)
axes.bar(fdf_graph_base['release_year'], fdf_graph_base['revenue'], label='Revenue')
axes.plot(fdf_graph_base['release_year'], fdf_graph_base['budget'], color='red',linewidth=3 , label='Budget')
axes.legend(loc=0)
```
