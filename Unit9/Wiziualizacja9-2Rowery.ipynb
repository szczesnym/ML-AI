```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import datetime as dt
```

```python
bike_df=pd.read_csv('daily-bike-share.csv')
```

```python
bike_df.head()
```

```python
bike_df.tail()
```

```python
sns.pairplot(bike_df[['temp', 'hum', 'windspeed', 'rentals']])
plt.show()
```

```python
sns.pairplot(bike_df[['season', 'temp', 'hum', 'windspeed', 'rentals']], hue='season', palette='bright')
plt.show()
```

```python
plt.figure(figsize=(9, 6))
sns.heatmap(bike_df[['temp', 'hum', 'windspeed', 'rentals']].corr(), vmin=-1, vmax=1, annot=True)
plt.show()
```
