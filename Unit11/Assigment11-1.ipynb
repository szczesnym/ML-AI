```python
import pandas as pd
import numpy as np
import plotly.subplots as subplots
import plotly.graph_objects as go
```

```python
kghm_stocks = pd.read_csv('11-1-data\\kgh_d.csv', index_col='Data', parse_dates=True)
cuprum_prices = pd.read_csv('11-1-data\\ca_c_f_d.csv', index_col='Data', parse_dates=True)

```

```python
kghm_stocks.head()
```

```python
cuprum_prices.head()
```

```python
fig = subplots.make_subplots(rows=3, cols=1, subplot_titles=['KGH stocks price ','Cuprum price'], shared_xaxes=True, vertical_spacing=0.0,
                             specs=[[{"type": "scatter"}],
                                    [{"type": "scatter"}],
                                    [{"type": "table"}]                                    ])
```

```python
fig.add_trace(go.Scatter(x=kghm_stocks.index, y=kghm_stocks['Zamkniecie'].tolist(), mode='lines', name='KGH stocks price'),row=1,col=1)
fig.add_trace(go.Scatter(x=cuprum_prices.index, y=cuprum_prices['Zamkniecie'].tolist(), mode='lines', name='Cuprum price'),row=2,col=1)
fig.add_trace(go.Table(header=dict(values=['Data', 'KGHM', 'Mied≈∫']),
                       cells=dict(
                           values=[
                            kghm_stocks.index.date.tolist(),
                            kghm_stocks['Zamkniecie'].tolist(),
                            cuprum_prices['Zamkniecie'].tolist()])
                       ), row=3,col=1
              )
fig.update_layout(
    height=1200,
    showlegend=False,
    title_text="Ceny akcji KGHM, miedzi i tabela",
)
fig.show()
```
